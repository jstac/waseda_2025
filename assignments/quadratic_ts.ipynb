# Ergodicity notes

Let's do some small experiments with the quadratic map.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return 4 * x * (1 - x)
```

```python
x_grid = np.linspace(0, 1, 200)
fig, ax = plt.subplots()
ax.plot(x_grid, g(x_grid))
plt.show()
```

Here's a time series.

```python
def generate_time_series(x_0, T=150):
    x = np.empty(T+1)
    x[0] = x_0
    for t in range(T):
        x[t+1] = g(x[t])
    return x
```

```python
fig, ax = plt.subplots()
ts = generate_time_series(0.55)
ax.plot(ts, '-o')
plt.show()
```
